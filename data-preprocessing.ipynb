```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/athlete_events.csv', header=0)
```

```python
df = df[df['Season'] == 'Summer']
df = df.drop(columns=['Games', 'Season'])
```

```python
df.head()
```

```python
regions = pd.read_csv('data/noc_regions.csv')
```

```python
regions.head()
```

```python
df = df.join(regions.set_index('NOC'), on='NOC')
df = df.drop(columns='notes')
```

```python
df.head()
```

```python
df['NOC'].unique()
```

```python
df[df['NOC'] == 'CHN'][['Year', 'Medal', 'NOC', 'region', 'Sport']].sort_values('Year')
```

```python
for country in df['NOC'].unique():
    df[df['NOC'] == country]
```

```python
participation = df.groupby('ID').aggregate({'Name': [lambda x: x[0:1], 'size']})
participation.columns = ['Name', 'Count']
```

```python
sport_participation = df.groupby('Sport').aggregate({'Name': lambda x: x.unique().shape[0]}).rename(columns={'Name': 'Athlete_Count'}).sort_values(by='Athlete_Count', ascending=False)
```

```python
participation.to_csv('data/athlete_participation.csv', index=False)
```

```python
sport_participation.to_csv('data/sport_participation.csv')
```

```python
sport_participation
```

```python
plt.barh(sport_participation.index[-10:], sport_participation['Athlete_Count'][-10:])
```

```python
medals = df.dropna(subset=['Medal'])
```

```python
athlete_medals = medals.groupby('Name').size()
```

```python
athlete_medals.to_csv('data/athlete_medals.csv', header=True)
```

```python
athlete_medals = athlete_medals.sort_values(ascending=False)
```
